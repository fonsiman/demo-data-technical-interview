# TOP 500 supercomputer mining - Challenge 1

# 1-A

In this challenge you will mine top500 website to retrieve information on the most powerfull supercomputer of the last years.

Your task is to scrape information for the top 500 supercomputers from 2014 to 2018. The information should include:

* **Rank**

* **Name**

* **Year**

* **Description**

* **Company**

* **Country**

* **Cores**

* **RMax (TFlops/s)**

* **RPeak (TFlops/s)**

* **Power (kW)**


The root url is the following:
https://www.top500.org/


You need to navigate the web in order to extract all information for 2500 supercomputers. As an example, the first urls are:

https://www.top500.org/lists/2014/11/?page=1


https://www.top500.org/list/2014/11/?page=2



Write code that do the following:
1. Look for the url to scrape.
2. Make request.
3. Return HTML string.
4. Parse and extract all features. Create a dictionary with these fields.
5. Return a list of dictionaries.



````python
# Your code here
import requests as rq
from bs4 import BeautifulSoup

rank = []
name = []
year = []
month = []
description = []
company = []
country = []
cores = []
rmax = []
rpeak = []
power = []
months = [6,11]

for c_year in range(2014,2019):
    for c_month in months:
        for page in range(1,6):
            print(page)

            url = f"https://www.top500.org/list/{c_year}/{c_month}/?page={page}"

            data = rq.get(url)

            html = BeautifulSoup(data.text, 'html.parser')
            table = html.select('table tr')[1:]

            for e in range(len(table)):
                rank.append(table[e].select("td:nth-child(1)")[0].text)
                name.append(table[e].select("td:nth-child(3) b")[0].text if len(table[e].select("td:nth-child(3) b")) == 1 \
                             else "None")
                year.append(c_year)
                month.append(c_month)
                description.append(table[e].select("td:nth-child(2) a")[0].text + " " + \
                    "".join(table[e].select("td:nth-child(3) a")[0].text.replace(name[e],"").strip().split(",")[:-1])[2:].partition(" ")[2])
                company.append(table[e].select("td:nth-child(3) br")[0].text)
                country.append(table[e].select("td:nth-child(2)")[0].find(text=True, recursive=False))
                cores.append(table[e].select("td:nth-child(4)")[0].text.replace(",",""))
                rmax.append(table[e].select("td:nth-child(5)")[0].text.replace(",",""))
                rpeak.append(table[e].select("td:nth-child(6)")[0].text.replace(",",""))
                power.append(table[e].select("td:nth-child(7)")[0].text.replace(",",""))

    
print_var = cores
print(len(print_var))
print(print_var)
````

````python
# Creamos el diccionario:

d = {
    "year" : year,
    "month" : month,
    "rank" : rank,
    "name" : name,
    "description" : description,
    "company" : company,
    "country" : country,
    "cores" : cores,
    "rmax" : rmax,
    "rpeak" : rpeak,
    "power" : power
}

print(cores)
````

````python
# Your code here

import getpass
import pymysql
from sqlalchemy import create_engine

mysql_domain = "localhost"
mysql_user = "root"
mysql_password = getpass.getpass()
mysql_db = "prueba_tecnica"

dbdata = [mysql_user, mysql_password, mysql_domain, mysql_db]
connection_url = 'mysql+pymysql://{}:{}@{}/{}'.format(*dbdata)
engine = create_engine(connection_url)
print("Connected to DB!")
````

````python
sqlquery = """
CREATE TABLE top_500(
        ranking int NOT NULL AUTO_INCREMENT,
        name text, 
        year int,
        month int,
        description text,
        company text,
        country text, 
        cores int, 
        rmax float,
        rpeak float,
        power int,
        PRIMARY KEY (id));
"""
from sqlalchemy import MetaData, Integer, Table, Column, text
from sqlalchemy.dialects.mysql import FLOAT, TEXT

m = MetaData()
t = Table('top500', m,
        Column('ranking', Integer),
        Column('name', TEXT),
        Column('year', Integer),
        Column('month', Integer),
        Column('description', TEXT),
        Column('company', TEXT),
        Column('country', TEXT),
        Column('cores', Integer),
        Column('rmax', FLOAT(precision=8, scale=1)),
        Column('rpeak', FLOAT(precision=8, scale=1)),
        Column('power', Integer)
    )

m.create_all(engine)
````

````python
from sqlalchemy.sql.expression import insert

for e in range(len(rank)):
    if power[e] != "":
        ins = t.insert().values(ranking=rank[e], name=name[e], year=year[e], month=month[e], description=description[e],\
                                company=company[e], country=country[e], cores=cores[e], rmax=rmax[e], rpeak=rpeak[e], \
                                power=power[e])
    else:
        ins = t.insert().values(ranking=rank[e], name=name[e], year=year[e], month=month[e], description=description[e],\
                                company=company[e], country=country[e], cores=cores[e], rmax=rmax[e], rpeak=rpeak[e])
    engine.execute(ins)
````
